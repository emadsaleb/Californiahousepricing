# load California House Price Data Set

```python
from sklearn.datasets import fetch_california_housing
```

```python
california= fetch_california_housing()
```

# preparing the Data set

```python
dataset = pd.DataFrame(california.data , columns = california.feature_names)
```

```python
dataset.head()
```

```python
dataset.tail()
```

```python
dataset['Price'] = california.target
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

# Checking the missing values

```python
dataset.isnull().sum()
```

```python
# Correlation
dataset.corr()
```

```python
plt.scatter(dataset['HouseAge'] , dataset['Price'])
plt.xlabel('HouseAge')
plt.ylabel('Price')
plt.show()
```

```python
plt.scatter(dataset['AveRooms'] , dataset['Price'])
plt.xlabel('AveRooms')
plt.ylabel('Price')
plt.show()
```

```python
plt.scatter(dataset['Population'] , dataset['Price'])
plt.xlabel('Population')
plt.ylabel('Price')
plt.show()
```

```python
# Independant and Dependent features
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
X.head()
```

```python
y
```

# training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.3 , random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
# standarize the dataset
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
import pickle

pickle.dump(scaler, open("Scaling.pkl", "wb"))
```

```python
X_train
```

```python
X_test
```

# Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train , y_train)
```

```python
## print the coefficients and the intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
regression.get_params()
```

```python
# Prediction
reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

# Plotting

```python
plt.scatter(x = y_test , y = reg_pred)
```

```python
## residuals
residuals= y_test - reg_pred
```

```python
residuals
```

```python
plt.scatter(reg_pred , residuals)
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error
print(mean_absolute_error(y_test , reg_pred))
print(mean_squared_error(y_test , reg_pred))
print(np.sqrt(mean_squared_error(y_test , reg_pred)))
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test , reg_pred)
```

```python
score
```

```python
#display adjusted R-squared
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
